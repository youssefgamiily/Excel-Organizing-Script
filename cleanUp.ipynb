```python
import openpyxl
import json

class Item:
    def __init__(self, inventory_code, auction_code, name, selling_price, wrapping_cost, status, item_flag):
        self.inventory_code = inventory_code
        self.auction_code= auction_code
        self.name = name
        self.selling_price = selling_price
        self.wrapping_cost = wrapping_cost
        self.status = status
        self.flag = item_flag

    def to_dict(self):
        return {
            "inventory_code": self.inventory_code,
            "auction_code": self.auction_code,
            "name": self.name,
            "selling_price": self.selling_price,
            "wrapping_cost": self.wrapping_cost,
            "status": self.status,
            "flag": self.flag
        }
    def to_json(self):
        return json.dumps(self.to_dict(), indent=2)


class Buyer:
    def __init__(self, name, city, address, number, status):
        self.name = name
        self.city = city
        self.address = address
        self.number = number
        self.status = status

    def to_dict(self):
        return {
            "name": self.name,
            "city": self.city,
            "address": self.address,
            "number": self.number,
            "status": self.status
        }
```

```python
# DFs p1
import pandas as pd
import json

# Load Excel workbook using pandas
xl_file_path = './Auctions Sales Recording 11 Oct 233.xlsx'
xl = pd.ExcelFile(xl_file_path)

# Dictionaries to store buyer and item data
buyer_dict = {}
item_dict = {}

# Iterate over each sheet
for sheet_name in xl.sheet_names:
    df = pd.read_excel(xl, sheet_name)

    # Filter rows based on conditions
    is_auction = df.columns[0].isnumeric()
    auction_num = int(df.columns[0]) if is_auction else 9999

    if is_auction and auction_num > 1760:
        # Filter columns to match your data
        df_filtered = df.iloc[1:100, [0, 1, 2, 3, 8, 9, 10]]

        # Save the filtered data as a CSV file
        csv_filename = f'{sheet_name}_filtered.csv'
        df_filtered.to_csv(csv_filename, index=False, encoding='utf-8')

        # Create a buyer dataframe
        buyer_df = df_filtered.iloc[:, [3, 4, 5, 6, 10]]
        print (buyer_df)
        # Create an item dataframe
        item_df = df_filtered.iloc[:, [0, 1, 2, 3, 8, 9]]

        # Save the buyer dataframe to a dictionary
        for _, row in buyer_df.iterrows():
            buyer_name = row[0]
            if buyer_name not in buyer_dict:
                buyer_dict[buyer_name] = {
                    'city': row[1],
                    'address': row[2],
                    'number': row[3],
                    'status': row[4],
                    'auctions': [auction_num]
                }
            else:
                buyer_dict[buyer_name]['auctions'].append(auction_num)

        # Save the item dataframe to a dictionary
        for _, row in item_df.iterrows():
            item_code = row[0]
            if item_code not in item_dict:
                item_dict[item_code] = {
                    'auction_code': auction_num,
                    'name': row[2],
                    'selling_price': row[3],
                    'wrapping_cost': row[8],
                    'status': row[9],
                    'item_flag': row[11]
                }
            else:
                item_dict[item_code]['auction_code'] = auction_num

# Save buyer and item dictionaries to JSON files
buyer_json_filename = "buyer_data.json"
item_json_filename = "item_data.json"

with open(buyer_json_filename, mode='w', encoding='utf-8') as buyer_json_file:
    json.dump(buyer_dict, buyer_json_file, indent=2)

with open(item_json_filename, mode='w', encoding='utf-8') as item_json_file:
    json.dump(item_dict, item_json_file, indent=2)

print(f"Buyer data has been dumped into '{buyer_json_filename}'.")
print(f"Item data has been dumped into '{item_json_filename}'.")

#DFs P2

# Load the buyer and item dataframes from the JSON files
buyer_df = pd.read_json('buyer_data.json', orient='index')
item_df = pd.read_json('item_data.json', orient='index')
```

```python
import pandas as pd
import json

# Load the JSON file
with open('buyer_items_data.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Initialize empty lists to store data
buyer_data = []
item_data = []

# Iterate through the JSON data
for buyer_name, buyer_info in data.items():
    buyer_details = buyer_info[0]
    items = buyer_info[1]

    # Append buyer details to buyer_data list
    buyer_row = {
        'buyer_name': buyer_details['name'],
        'city': buyer_details['city'],
        'address': buyer_details['address'],
        'number': buyer_details['number'],
        'status': buyer_details['status']
    }
    buyer_data.append(buyer_row)

    # Append item details to item_data list
    for item in items:
        item_row = {
            'buyer_name': buyer_details['name'],
            'inventory_code': item['inventory_code'],
            'auction_code': item['auction_code'],
            'name': item['name'],
            'selling_price': item['selling_price'],
            'wrapping_cost': item['wrapping_cost'],
            'status': item['status'],
            'item_flag': item['flag']
        }
        item_data.append(item_row)

# Create DataFrames from the lists
buyer_df = pd.DataFrame(buyer_data)
item_df = pd.DataFrame(item_data)

# Merge buyer_df and item_df on the 'buyer_name' column
merged_df = pd.merge(buyer_df, item_df, on='buyer_name', how='inner')

# Convert 'selling_price' column to numeric
merged_df['selling_price'] = pd.to_numeric(merged_df['selling_price'], errors='coerce')

# Create a mask for buyers meeting the specified conditions
private_shipping_mask = (
    (merged_df['city'].str.lower().isin(['alex', 'cairo'])) &
    ((merged_df['item_flag'].notnull()) | (merged_df.groupby('buyer_name')['selling_price'].transform('sum') > 1000))
)
AlexPriv_shipping_mask = (merged_df['city'].str.lower().isin(['alex'])) & private_shipping_mask
CairoPriv_shipping_mask = (merged_df['city'].str.lower().isin(['cairo'])) & private_shipping_mask

# Create Private Shipping and Outsourced DataFrames
private_shipping_df = merged_df[private_shipping_mask].copy()
outsourced_df = merged_df[~private_shipping_mask].copy()
AlexPriv_shipping_df = merged_df[AlexPriv_shipping_mask].copy()
CairoPriv_shipping_df = merged_df[CairoPriv_shipping_mask].copy()

# Display the DataFrames

print("merged priv shipping df : ")
print(private_shipping_df)

print("cairo priv shipping df : ")
print(CairoPriv_shipping_df)

print("alex priv shipping df : ")
print(AlexPriv_shipping_df)

# Save the DataFrames to separate sheets in the same Excel file using openpyxl
excel_file_path = 'output.xlsx'
with pd.ExcelWriter(excel_file_path, engine='openpyxl') as writer:
    merged_df.to_excel(writer, sheet_name='Merged', index=False)
    private_shipping_df.to_excel(writer, sheet_name='Private_Shipping', index=False)
    AlexPriv_shipping_df.to_excel(writer, sheet_name='Alex Private_Shipping', index=False)
    CairoPriv_shipping_df.to_excel(writer, sheet_name='Cairo Private_Shipping', index=False)
    outsourced_df.to_excel(writer, sheet_name='Outsourced', index=False)
```

```python
import openpyxl

# Load the Excel workbook
wb = openpyxl.load_workbook('output.xlsx')

# Create new sheets for categories
categories = ['Cairo Priv', 'Cairo Outsource', 'Alex Priv', 'Alex Outsource', 'Others']
for category in categories:
    wb.create_sheet(category)

# Get the original sheet
original_sheet = wb['Sheet1']  # Assuming your original sheet is named 'Sheet1'

# Iterate over rows
for row in original_sheet.iter_rows(min_row=2, max_row=original_sheet.max_row, min_col=1, max_col=original_sheet.max_column):
    buyer_name = row[0].value

    # Sum numeric values, skipping non-numeric values
    total_sum = sum(float(cell.value) if cell.value is not None and isinstance(cell.value, (int, float, str)) and cell.value.isdigit() else 0 for cell in row[2:])
    
    has_flag = any(flag.value for flag in row[3:])
    is_alex_cairo = 'Alex' in buyer_name or 'Cairo' in buyer_name

    # Check conditions
    if total_sum > 1000 and (has_flag or is_alex_cairo):
        print(f"Condition satisfied for buyer {buyer_name}")

        # Determine category
        if 'Alex' in buyer_name:
            category = 'Alex Priv' if has_flag else 'Alex Outsource'
        elif 'Cairo' in buyer_name:
            category = 'Cairo Priv' if has_flag else 'Cairo Outsource'
        else:
            category = 'Others'

        # Add the row to the corresponding category sheet
        category_sheet = wb[category]
        category_sheet.append([cell.value for cell in row])

# Save the modified workbook
wb.save('modified_output.xlsx')
```
